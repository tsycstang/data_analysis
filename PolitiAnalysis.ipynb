```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
path = '/content/2014data.csv'
raw_head = ['serial', 'time', 'nan1']
qlist = ['q01', 'q02', 'q03', 'q04', 'q05', 'q06', 'q07', 'q08', 'q09', 'q10', 
         'q11', 'q12', 'q13', 'q14', 'q15', 'q16', 'q17', 'q18', 'q19', 'q20', 
         'q21', 'q22', 'q23', 'q24', 'q25', 'q26', 'q27', 'q28', 'q29', 'q30', 
         'q31', 'q32', 'q33', 'q34', 'q35', 'q36', 'q37', 'q38', 'q39', 'q40', 
         'q41', 'q42', 'q43', 'q44', 'q45', 'q46', 'q47', 'q48', 'q49', 'q50']
personal_info = ['gender', 'birth_year', 'income', 'education']
columns = raw_head + qlist + personal_info
data_raw = pd.read_csv(path, 
                       names=columns)
questions = {}
for i in qlist:
    questions[i] = data_raw[i][0]
data = data_raw.drop(raw_head, axis=1).drop([0])
```

```
for i in qlist:
    print(i, ': ', questions[i])
    print(data[i].value_counts(), '\n')
```

```
quantitized_data = data.copy()
quantitize_dict = {'同意': 0.5, 
                   '强烈同意': 1.5, 
                   '反对': -0.5, 
                   '强烈反对': -1.5}
quantitized_data = quantitized_data.replace(quantitize_dict)
```

```
quan_data_q = quantitized_data.copy()
quan_data_q = quan_data_q.drop(['gender', 'birth_year', 'income', 'education'], axis=1)
```

```
quan_data_q.describe()
```

```
corr_matrix = quan_data_q.corr()
```

```
corr_matrix
```

```
corr_mat_mod = corr_matrix.copy()
for i in qlist:
    corr_mat_mod[i][i] = 0.
corr_mat_mod.describe()
```

```
corr_dict = {}
for i in qlist:
    for j in qlist:
        if i != j and (i, j) not in corr_dict and (j, i) not in corr_dict:
            corr_dict[(i, j)] = corr_matrix[i][j]
corr_S = pd.Series(corr_dict)
```

```
corr_df = pd.DataFrame(corr_S, columns=['corr'])
```

```
corr_df.sort_values(by='corr')
```

```
corr_maxmin = {}
corr_info = corr_mat_mod.describe()
for i in qlist:
    corr_maxmin[i] = corr_info[i]['max'] - corr_info[i]['min']
```

```
corr_mm_df = pd.DataFrame(pd.Series(corr_maxmin), columns=['corr_range'])
```

```
corr_mm_df.sort_values(by='corr_range')
```
